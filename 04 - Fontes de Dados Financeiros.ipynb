Índice Bovespa: ^BVSP

```python
# Baixando cotações dos últimos 12 meses de PETR4
petr4 = yf.download('PETR4.SA', period='1y')
```

```python
petr4
```

```python
# Baixando dados intraday, minuto a minuto de VALE3.SA, dos últimos 5 pregões

# Valid Intervals: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
# Valid Periods: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max

vale3 = yf.download('VALE3.SA', interval='1m', period='5d')
```

```python
vale3
```

```python
# Baixando várias empresas de uma só vez. Método 1 (Tickers separados por espaço)
yf.download('WEGE3.SA ITSA4.SA')
```

```python
# Baixando várias empresas de uma só vez. Método 2 (Passando uma lista de tickers)
yf.download(['WEGE3.SA', 'ITSA4.SA'])
```

```python
# Alterando o método de agrupamento
# Group by 'ticker' ou 'column' (default)
yf.download(['WEGE3.SA', 'ITSA4.SA'], group_by='ticker')
```

### Obtendo e Analisando os Dados

```python
ALPHAVANTAGE_API_KEY = ''
```

```python
# Criando o objeto ts do tipo TimeSeries
ts = TimeSeries(key=ALPHAVANTAGE_API_KEY, output_format='pandas')
```

```python
# Pesquisando o código dos ativos com get_symbol_search()
ts.get_symbol_search('BOVA')
```

```python
# Obtendo os dados semanais do IBOV usando get_weekly()
ts.get_weekly('IBOV.SAO')
```

```python
# Obtendo os dados e metadados diários do IBOV usando get_daily() com outputsize='full'
# dados, metadados = ts.get_daily('IBOV.SAO', outputsize='full')

# NÃO FUNCIONA MAIS - SOMENTE COM ACESSO PREMIUM
```

```python
# Obtendo os dados e metadados diários do IBOV usando get_daily()
dados, metadados = ts.get_daily('IBOV.SAO')
```
