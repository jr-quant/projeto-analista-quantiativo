```python
# Ler arquivo CSV com Pandas e definir a coluna 0 como índice
URL = 'https://media.githubusercontent.com/media/codigoquant/data/refs/heads/main/indicadores_fundamentalistas.csv'

empresas = pd.read_csv(URL, index_col=0)

# Exiba o conteúdo do arquivo
empresas
```

```python
# Criando filtro de empresas com liquidez maior que 1B
filtro = empresas['Liq.2meses'] > 1000000000

# Listando as empresas com o filtro
empresas[filtro]
```

```python
# Ordenando empresas por menores P/L
empresas.sort_values('P/L')[['Papel', 'P/L']]
```

```python
# Ordenando empresas por maiores Dividend Yield
empresas.sort_values('Div.Yield', ascending=False)[['Papel', 'Div.Yield']]
```

```python
# Listando as 5 empresas com maior ROIC
empresas.sort_values('ROIC', ascending=False).head()[['Papel', 'ROIC']]
```

```python
# Listar as 5 empresas com menor EV/EBIT, excluindo EV/EBIT negativo
empresas[empresas['EV/EBIT'] > 0].sort_values('EV/EBIT').head()[['Papel', 'EV/EBIT']]
```

```python
# Colunas Papel e P/L das 10 empresas com menor P/L não negativo, e com Liq.2meses entre R$1 mi e R$ 10 mi
empresas[(empresas['Liq.2meses'] > 1000000) & (empresas['Liq.2meses'] < 10000000) & (empresas['P/L'] > 0)][['Papel', 'P/L']].sort_values('P/L')[:10]
```
