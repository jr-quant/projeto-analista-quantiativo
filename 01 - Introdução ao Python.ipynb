## Valor Futuro (FV)

Qual seria o Valor Futuro (FV) se eu investisse R$ 1.000,00 por 1 ano, com um retorno de 5% ao ano?

$$
FV = PV(1+r)^n
$$

FV = Valor Futuro (Future Value)

PV = Valor Presente (Present Value)

r = taxa (rate) de juros

n = número (number) de períodos


```python
PV = 1000
r = 0.05
n = 1

FV = PV * ((1 + r) ** n)
FV
```

## Valor Presente (PV)

Qual seria o Valor Presente (PV), se eu tivesse que descontar R$ 1.050,00 a uma taxa anual de 5% por um período de 1 ano?

$$
PV = \frac{FV}{(1+r)^n}
$$

FV = Valor Futuro (Future Value)

PV = Valor Presente (Present Value)

r = taxa (rate) de juros

n = número (number) de períodos

```python
PV = 1050
r = 0.05
n = 1

PV = FV / ((1 + r) ** n)
PV
```
